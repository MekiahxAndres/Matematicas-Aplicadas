#**Coste Marginal**

```python
import sympy as sp


x = sp.symbols('x')

P = 5.00 - 0.002 * x
C = 3.00 + 1.10 * x

R = P * x

Cm = sp.diff(C, x)
Rm = sp.diff(R, x)

U = R - C
Um = sp.diff(U, x)

solucion_x = sp.solve(Um, x)

print("Funciones dadas:")
print(f"P(x) = {P}")
print(f"C(x) = {C}")
print(f"R(x) = {R}")
print(f"U(x) = {U}")

print("\nDerivadas:")
print(f"Costo marginal, Cm(x) = {Cm}")
print(f"Ingreso marginal, Rm(x) = {Rm}")
print(f"Utilidad marginal, Um(x) = {Um}")

print("\nNivel de producción que maximiza la utilidad:")
print(f"x = {solucion_x[0]} unidades")
```

## Solución

1). **Definición de Funciones**

   Definimos las funciones de **precio** \( P(x) \) y **costo** \( C(x) \), ambas en función de la cantidad \( x \):

   - **Precio**: \( P(x) = 5.00 - 0.002x \)
   - **Costo**: \( C(x) = 3.00 + 1.10x \)

   Estas ecuaciones reflejan cómo el precio y el costo cambian con la cantidad de unidades \( x \) producidas o vendidas.




2). **Cálculo del Ingreso**

   El **ingreso** \( R(x) \) se calcula multiplicando el precio \( P(x) \) por la cantidad \( x \):

   $$
   R(x) = P(x) \cdot x = (5.00 - 0.002x) \cdot x = 5.00x - 0.002x^2
   $$

   Esto representa el ingreso total obtenido al vender \( x \) unidades.



3). **Derivadas para los Marginales**

   - **Costo Marginal \( C_m(x) \)**: Se obtiene derivando la función de costo \( C(x) \) respecto a \( x \):

     $$
     C_m(x) = \frac{dC}{dx} = 1.10
     $$

     Esto indica que el costo por unidad adicional es constante en \( 1.10 \) dólares.

   - **Ingreso Marginal \( R_m(x) \)**: Derivamos la función de ingreso \( R(x) \) respecto a \( x \):

     $$
     R_m(x) = \frac{dR}{dx} = 5.00 - 0.004x
     $$

     Este es el ingreso adicional generado por cada unidad extra vendida.

   - **Utilidad Marginal \( U_m(x) \)**: Primero, definimos la **utilidad** \( U(x) \) como la diferencia entre ingreso y costo:

     $$
     U(x) = R(x) - C(x) = (5.00x - 0.002x^2) - (3.00 + 1.10x)
     $$

     Simplificando:

     $$
     U(x) = -0.002x^2 + 3.90x - 3.00
     $$

     Luego, derivamos \( U(x) \) respecto a \( x \) para obtener la utilidad marginal:

     $$
     U_m(x) = \frac{dU}{dx} = -0.004x + 3.90
     $$




4). **Maximización de la Utilidad**

   Para maximizar la utilidad, igualamos la utilidad marginal a cero y resolvemos para \( x \):

   $$
   -0.004x + 3.90 = 0
   $$

   Resolviendo esta ecuación:

   $$
   x = \frac{3.90}{0.004} = 975
   $$

   Esto significa que producir **975 unidades** maximiza la utilidad total.



5). **Resumen de Resultados**

   - **Ingreso marginal**: \( R_m(x) = 5.00 - 0.004x \)
   - **Costo marginal**: \( C_m(x) = 1.10 \)
   - **Utilidad marginal**: \( U_m(x) = -0.004x + 3.90 \)
   - **Nivel de producción óptimo**: \( x = 975 \) unidades

Cada derivada representa el cambio en el ingreso, costo y utilidad cuando se incrementa en una unidad la cantidad producida. El nivel de producción que maximiza la utilidad ocurre cuando el ingreso marginal es igual al costo marginal.


#**Método de Newton**
1.Analice el siguiente codigo. Explique cada paso en el codigo anterior y compruebe

```python
#Definicion de la variable newton para encontrar una raiz
def newton(fun, der, x_n, epsilon=1e-6, steps=50):
    #inicia un bucle de iteracion
    for n in range(steps + 1):
        # Evaluación de la función para ver si el resultado es válido
        f_x = fun(x_n)
        if abs(f_x) < epsilon:
            return x_n

        # Evaluación de la derivada
        d_f = der(x_n)
        if d_f == 0:
            print('Error la derivada es cero')
            return None

        # Estimación del siguiente punto
        x_n = x_n - f_x / d_f
        #Retorno de la funcion
        return None
```

```python
import math

# Definición del método de Newton
def newton(fun, der, x_n, epsilon=1e-6, steps=50):
    for n in range(steps + 1):
        f_x = fun(x_n)
        if abs(f_x) < epsilon:
            return x_n

        d_f = der(x_n)
        if d_f == 0:
            print('Error la derivada es cero')
            return None

        x_n = x_n - f_x / d_f

    return None  # Fuera del bucle
```

### 1. Ecuación: x^5 - 6x^3 + 2x - 2 = 0

```python
f1 = lambda x: x**5 - 6*x**3 + 2*x - 2
g1 = lambda x: 5*x**4 - 18*x**2 + 2
sol1 = newton(f1, g1, x_n=1.5, epsilon=0.0001)
print("Solución 1:", sol1)
```

### 2. Ecuación: sin(x) - x^2 = 0


```python
f2 = lambda x: math.sin(x) - x**2
g2 = lambda x: math.cos(x) - 2*x
sol2 = newton(f2, g2, x_n=0.5, epsilon=0.0001)
print("Solución 2:", sol2)
```

### 3. Ecuación: cos(t) - t^2 = 0


```python
f3 = lambda t: math.cos(t) - t**2
g3 = lambda t: -math.sin(t) - 2*t
sol3 = newton(f3, g3, x_n=0.5, epsilon=0.0001)
print("Solución 3:", sol3)
```

### 4. Ecuación: 2*sin(x) - sqrt(x + 1) = 0


```python
f4 = lambda x: 2 * math.sin(x) - math.sqrt(x + 1)
g4 = lambda x: 2 * math.cos(x) - 0.5 / math.sqrt(x + 1)
sol4 = newton(f4, g4, x_n=0.5, epsilon=0.0001)
print("Solución 4:", sol4)
```

```python
import matplotlib.pyplot as plt
import numpy as np


def f(x, a):
    return x**2 + a*x**4


def f_prima(x, a):
    return 2*x + 4*a*x**3


def f_segunda(x, a):
    return 2 + 12*a*x**2


a = 2


x = np.linspace(-2, 2, 100)


y = f(x, a)
y_prima = f_prima(x, a)
y_segunda = f_segunda(x, a)


fig, axs = plt.subplots(3, 1, figsize=(10, 10))


axs[0].plot(x, y)
axs[0].set_title('f(x)')
axs[0].grid(True)


axs[1].plot(x, y_prima)
axs[1].set_title("f'(x)")
axs[1].grid(True)


axs[2].plot(x, y_segunda)
axs[2].set_title("f''(x)")
axs[2].grid(True)


plt.show()
```
