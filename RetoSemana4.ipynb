### 1. Serie Fubonacci

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
n = int(input("Ingresa el número de términos de la serie de Fibonacci que deseas generar: "))
fibonacci = [0, 1]

for i in range(2, n):
    siguiente_fib = fibonacci[i-1] + fibonacci[i-2]
    fibonacci.append(siguiente_fib)

print(f"Los primeros {n} números de la serie de Fibonacci son: {fibonacci}")

plt.plot(fibonacci, marker='o', linestyle='-', color='b')
plt.title(f"Serie de Fibonacci - Primeros {n} términos")
plt.xlabel("Índice")
plt.ylabel("Valor Fibonacci")
plt.grid(True)
plt.show()
```

###7.Movimiento parabólico:

```python
def proyectil(v0, theta, g=9.81):

    theta = np.radians(theta)

    vx0 = v0 * np.cos(theta)
    vy0 = v0 * np.sin(theta)

    t_vuelo = 2 * vy0 / g

    t = np.linspace(0, t_vuelo, 100)


    x = vx0 * t
    y = vy0 * t - 0.5 * g * t**2

    return t, x, y

v0 = 30
angulos = [30, 35, 40, 45, 60, 65]

plt.figure(figsize=(8, 6))

for angulo in angulos:
    t, x, y = proyectil(v0, angulo)
    plt.plot(x, y, label=f"Ángulo = {angulo}°")

plt.title("Movimiento de un proyectil")
plt.xlabel("Distancia horizontal en metros")
plt.ylabel("Altura en metros")
plt.grid(True)
plt.show()
```

### 9.Fractales con Python

```python
import matplotlib.pyplot as plt
import random

# Coordenadas iniciales del triángulo
x = [0, 1, 0.5]
y = [0, 0, 0.8]

# Punto inicial
px = 0.5
py = 0.5

# Listas para almacenar los puntos generados
puntosx = []
puntosy = []

# Ciclo para generar el fractal
n = 0
while n < 100000:  # Se generarán 100,000 puntos
    # Escoge un vértice al azar
    vertice = random.randint(0, 2)

    # Mueve el punto a la mitad de camino hacia el vértice escogido
    px = (px + x[vertice]) / 2
    py = (py + y[vertice]) / 2

    # Almacena el punto generado
    puntosx.append(px)
    puntosy.append(py)

    n += 1

# Visualización del fractal
plt.scatter(puntosx, puntosy, s=0.1,)
plt.axis('equal')
plt.title("Fractal de Sierpinski")
plt.grid()
plt.show()
```
